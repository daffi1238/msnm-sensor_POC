```python
import numpy as np
from datetime import datetime
import time
from scipy.stats import norm
import numpy.linalg as linalg


import matplotlib.pyplot as plt
%matplotlib inline
```

```python
rng = np.random.RandomState(0)
rng
```

```python
W = rng.randn(1000, 2)
W
```

```python
plt.scatter(W[:,0], W[:,1])
plt.axis('equal')
plt.grid(True)
```

```python
#Una matriz aleatoria se genera y multiplica para cambiarle la forma a la distribución
#de los puntos y ver más claramente el funcionamiento del PCA
A = rng.rand(2,2)
A
```

```python
X = np.dot(W,A)
```

```python
plt.scatter(X[:,0], X[:,1])
plt.axis('equal')
plt.grid(True)
```

```python
np.var(X[:,0])
```

```python
np.var(X[:,1])
```

```python
#Covariance Matrix
C = np.cov(X.T)
C
#Para encontrar el primer componente principal (PC1) debemos utilizar
#los autovalores y autovectores de la matriz de covarianza
```

```python
#Calculate the eigevalues and eigenvectors
e, V = np.linalg.eig(C)
```

```python
#Estos son los autovalores de la matriz covarianza
e
```

```python
#Cada columna es un autovector de la matriz de covarianza
print V
```

```python
u = V[:,1] #Autovector correspondiente al mayor de los autovalores

u.shape
```

```python
Z = np.dot(X,u)
Z
```

```python
np.var(Z) #Será un valor (teóricament eigual) al autovalor del cual estamos 
#partiendo, es decir el segundo elemento de la matriz e
```

```python
#Debe ser 1
np.dot(u,u)
```

```python
Y = np.outer(Z,u)
Y.shape
```

```python
plt.scatter(X[:,0], X[:,1])
plt.scatter(Y[:,0], Y[:,1])
plt.axis('equal')
plt.grid(True)

#Hasta aqui tenemos que hemos calculado la primera componente principal en base
#a la matriz de covarianza
```

## Ampliando a 3 dimensiones

```python
import numpy as np
from datetime import datetime
import time
from scipy.stats import norm

import matplotlib.pyplot as plt
%matplotlib inline

#Para el 3D
from matplotlib import pyplot
from mpl_toolkits.mplot3d import Axes3D
```

```python
rng = np.random.RandomState(0)
rng

W = rng.randn(100, 3)
W
```

```python



#Plot en 3 dimensiones para ver 
fig = pyplot.figure()
ax = Axes3D(fig)

sequence_containing_x_vals = W[:,0]
sequence_containing_y_vals = W[:,1]
sequence_containing_z_vals = W[:,2]



ax.scatter(sequence_containing_x_vals, sequence_containing_y_vals, sequence_containing_z_vals)
pyplot.show()
```

```python
#Debido a la uniformidad de la muestra para hacer más fácil el entendimiento
#transformamos la matriz con un producto vectorial
A = rng.rand(3,3)

X = np.dot(W,A)
#Plot en 3 dimensiones para ver 
fig = pyplot.figure()
ax = Axes3D(fig)

sequence_containing_x_vals = X[:,0]
sequence_containing_y_vals = X[:,1]
sequence_containing_z_vals = X[:,2]



ax.scatter(sequence_containing_x_vals, sequence_containing_y_vals, sequence_containing_z_vals)
pyplot.show()
```

```python
#Covariance Matrix
C = np.cov(X.T)
C
#Para encontrar el primer componente principal (PC1) debemos utilizar
#los autovalores y autovectores de la matriz de covarianza
```

```python
#Calculate the eigevalues and eigenvectors de la amtriz de covarianza
e, V = np.linalg.eig(C)
```

```python
u = V[:,2] #Autovector correspondiente al mayor d elos autovalores

#¿Qué es Z? La proyección de los datos en su autovector asociado al mayor de los
#autovalores
Z = np.dot(X,u)
Z.shape


u2 = V[:,1] #Autovector correspondiente al mayor d elos autovalores

#¿Qué es Z? La proyección de los datos en su autovector asociado al mayor de los
#autovalores
Z2 = np.dot(X,u2)
Z2.shape
```

```python
Y = np.outer(Z,u)
Y

Y2 = np.outer(Z2,u2)
```

```python
#Plot en 3 dimensiones para ver 
fig = pyplot.figure()
ax = Axes3D(fig)

sequence_containing_x_vals = X[:,0]
sequence_containing_y_vals = X[:,1]
sequence_containing_z_vals = X[:,2]


sequence_containing_x_tend = Y[:,0]
sequence_containing_y_tend = Y[:,1]
sequence_containing_z_tend = Y[:,2]

sequence_containing_x_tend2 = Y2[:,0]
sequence_containing_y_tend2 = Y2[:,1]
sequence_containing_z_tend2 = Y2[:,2]



#Mostramos los puntos en un plano 3d
ax.scatter(sequence_containing_x_vals, sequence_containing_y_vals, sequence_containing_z_vals)

#Y dibujamos la tendencia de dichos datos
ax.scatter(sequence_containing_x_tend, sequence_containing_y_tend, sequence_containing_z_tend)


#Y dibujamos la tendencia de dichos datos
ax.scatter(sequence_containing_x_tend2, sequence_containing_y_tend2, sequence_containing_z_tend2)


pyplot.show()
```

Arriba vemos la nube de puntos (azul) y sus dos mayores autovectores (naranja y verde respectivamente) siendo estos los espacios donde al proyectar obtendremos la máxima variabilidad y por lo tanto conservaremos la máxima información.

Abajo se muestran las proyecciones en ambos espacios.

```python
plt.scatter(Y[:,0], Y[:,1])
plt.axis('equal')
plt.grid(True)

plt.scatter(Y2[:,0], Y2[:,1])
plt.axis('equal')
plt.grid(True)
```

# Using Scikit-learn

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn import datasets
from sklearn.decomposition import PCA
```

```python
iris = datasets.load_iris()
```

```python
X = iris.data
y = iris.target
```

```python
#El objetivo es reducir el número de columnas de X de 4 a 2 utilizando PCA
X.shape
```

```python
y.shape
```

```python
#Creamos el pca con 2 componentes
pca = PCA(2)
```

```python
pca.fit(X)
```

```python
#Para ver las dos componentes que tenemos definidas
pca.components_
#Recordemos que estos son los dos primeros autovectores de la matriz de covarianza del
#conjunto de observaciones
```

```python
#Vamos a transformar X en nuestro nuevo espacio de componentes principales
Z = pca.transform(X)
```

```python
Z.shape
```

```python
#El siguiente paso es comprobar cuanta información es preservada en este nuevo modelo
#y cuanta hemos perdido
plt.scatter(Z[:,0], Z[:,1], c=y)
```

```python
digits = datasets.load_digits()
```

```python
digits.data.shape
#¿Por qué 64?
```

```python
#Cada dígito es una imagen de 8x8
digits.images.shape
```

```python
plt.imshow(digits.images[0], cmap='gray')
#imagen del 0
```

```python
digits.target.shape
```

```python
#¿Podemos reducir el número de variables?
```

```python
pca = PCA(2)
```

```python
pca.fit(digits.data)
```

```python
Z = pca.transform(digits.data)
```

```python
Z.shape
```

```python
plt.scatter(Z[:,0], Z[:,1], c=digits.target, cmap='tab10', alpha=0.7)
plt.colorbar()
plt.grid(True)
```

Hemos conseguido obtener de tan sólo 2 componentes principales la información que encontrábamos en 64 componentes iniciales, es una herramienta muy potente
